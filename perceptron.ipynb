## Criando um Perceptron simples com python base

```python
## Variáveis de entrada
entradas = [1, 7, 5]
```

```python
## Pesos iniciais, números aleatórios = w
pesos = [0.8, 0.1, 0]
```

```python
## Função soma
def somaFunction(entradas, pesos):
    s = 0
    for i in range(len(entradas)):
        print(entradas[i], "*", pesos[i],"=", entradas[i] * pesos[i])
        s+= entradas[i] * pesos[i]

    return s
```

```python
## Executando somaFunction com entradas e pesos
soma = somaFunction(entradas, pesos)
print('soma =',soma)
```

```python
## Função de ativação
def stepFunction(soma):
    if soma>=1:
        return 1
    else:
        return 0
```

```python
## Executando a função de ativação no resultado da soma
resultado = stepFunction(soma)
print("resultado =",resultado)
```

## Criando um Perceptron simples com NumPy

```python
## Importando numpy
import numpy as np
```

```python
## Recriando as variáveis como array numpy
entradas_np = np.array([-1, 7, 5])

pesos_np = np.array(pesos)
```

```python
## Recriando função soma para numpy
def somaFunction_np(entradas_np, pesos_np):
    ## Produto escalar de matrizes/vetores
    return entradas_np.dot(pesos_np)
```

```python
## Soma usando função numpy
soma_np = somaFunction_np(entradas_np, pesos_np)
print('soma =',soma_np)
```

```python
## Resultado da ativação na soma numpy
resultado_np = stepFunction(soma_np)
print("resultado =",resultado_np)
```

## Criando classificação para um operador AND usando taxa de aprendizagem e erro

```python
import numpy as np
```

```python
## Definindo matriz de entrada (X)
entradas_class = np.array([[0,0], [0,1], [1,0], [1,1]])

print(entradas_class)
```

```python
## Definindo classe (y) - Operador AND
saidas_class_and = np.array([0, 0, 0, 1])

print(saidas_class_and)
```

```python
## Definindo pesos iniciais
pesos_class = np.array([0.0, 0.0])

print(pesos_class)
```

```python
## Definindo taxaAprendizagem
taxaAprendizagem = 0.1
```

```python
## Definindo stepFunction
def stepFunction_class(soma):
    if soma>=1:
        return 1
    else:
        return 0
```

```python
## Definindo função soma com numpy
def somaFunction_class(entradas, pesos):
    return entradas.dot(pesos)
```

```python
## Definindo função de treino do modelo
def treino(X, y, w, lr):
    erroTotal = 1
    while erroTotal != 0:
        # Atualizando variáveis do laço while
        erroTotal = 0
        w = w + lr

        for i in range(len(y)):
            # Gerando variáveis de resultado do cálculo de ativação após o produto escalar
            soma_class = somaFunction_class(X[i], w)
            resultado_class = stepFunction_class(soma_class)
            # Se o resultado for diferente do previsto, adiciona ao erroTotal
            if resultado_class != y[i]:
                erro = y[i] - resultado_class
                erroTotal += erro

    # Cria vetor de previsão
    y_pred = []
    for i in range(len(y)):
        soma_class = somaFunction_class(X[i], w)
        y_pred = y_pred + [stepFunction_class(soma_class)]

    # Retorna vetor de pesos e vetor de previsão
    return w, np.array(y_pred)
```

```python
output_treino = treino(X=entradas_class, y=saidas_class_and, w=pesos_class, lr=taxaAprendizagem)
print("vetor de pesos ótimo =", output_treino[0], "\nsaídas previstas =", output_treino[1], "\nsaídas reais =", saidas_class_and)
```

## Criando classificação para um operador OR usando taxa de aprendizagem e erro

```python
## Alterando o vetor de valor verdade para um operador OR
saidas_class_or = np.array([0, 1, 1, 1])
```

```python
output_treino = treino(X=entradas_class, y=saidas_class_or, w=pesos_class, lr=taxaAprendizagem)
print("vetor de pesos ótimo =", output_treino[0], "\nsaídas previstas =", output_treino[1], "\nsaídas reais =", saidas_class_or)
```

## Criando classificação para um operador XOR usando taxa de aprendizagem e erro

```python
## Alterando o vetor de valor verdade para um operador XOR
saidas_class_xor = np.array([0, 1, 1, 0])
```
